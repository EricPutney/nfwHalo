```python
import numpy as np
import scipy
from scipy import integrate
from numpy.random import default_rng
from multiprocessing import Pool
import nfwFunctions as NFW
import inverseTransform
import time
import sys

from scipy import special
from scipy.interpolate import RectBivariateSpline

import matplotlib as mpl
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.rcParams['figure.autolayout'] = False
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 14
mpl.rc('axes',edgecolor='k')
plt.rcParams['xtick.color'] = 'k'
plt.rcParams['ytick.color'] = 'k'
```

```python
class tracerGenerator():
    def __init__(self, nTracers, rScaleNFW, rho0, delta, useParallel = True, numThreads = 1, usePlummer = False, plummerMass = .5*10**10, rScalePlummer = 3*10**3):
        """
        Generates tracers from a pure NFW Halo or a Plummer cluster with NFW velocities. This init just stores some of the arguments passed in and computes some important/handy quantities.
        
        To actually generate the tracers, instantiate this class and then call the genTracers() method with whatever parameters you want.
        
        Properties of a tracerGenerator object:
            nTracers - number of tracer stars.
            rScaleNFW - often called "a" in the literature, this is the scale radius for the NFW profile.
            rho0 - scale density for NFW profile.
            delta - NFW cutoff scale, usually 200.
            c - the NFW halo concentration i.e. rDelta/rScaleNFW.
            rDelta - the radius at which we cut off tracer generation.
            mHalo - mass of the NFW halo.
            virialSpeed - virial speed of the NFW halo.
            plummerMass - If usePlummer == True, this controls the total mass of the plummer cluster of tracers
            rScalePlummer - If usePlummer == True, this controls the scale radius of the plummer cluster of tracers
        
        Properties of generated tracers:
            radii - radii of generated tracers (run genTracers first)
            normRadii - radii of generated tracers normalized to the cutoff radius rDelta (run genTracers first)
            speeds - speeds of the generated tracers (run genTracers first)
            coordAngles - the spherical coordinate angles (theta, phi) for each tracer in position space (run genTracers first)
            velocityAngles - the spherical coordinate angles (theta, phi) for each tracer in velocity space (run genTracers first)
            sphericalPhaseSpacePoints - contains the 6D phase space of all tracers in spherical coordinates
            cartesianPhaseSpacePoints - contains the 6D phase space of all tracers in cartesian coordinates
        """
        self.nTracers = nTracers
        self.rScaleNFW = rScaleNFW # tracerAttributes[0]
        self.rho0 = rho0 # tracerAttributes[1]
        self.delta = delta
        self.rhoCritical = 1.26*10**-7
        rhoNorm = self.rho0/(self.delta*self.rhoCritical)
        cShortcut = ((3.*np.sqrt(81.*rhoNorm**2. + 12.*rhoNorm) + 27.*rhoNorm + 2.)/2.)**(1./3.)
        self.c = ( cShortcut + 1./cShortcut - 2 )/3.
        self.rDelta = self.c*self.rScaleNFW # tracerAttributes[2]
        self.G = 4.3009*10**-3 # Newton's constant, (pc/solar mass) * (km/second)^2.
        self.mHalo = (4*np.pi*self.rho0*self.rScaleNFW**3)*NFW.gFcn(self.c) # Should be in units of solar masses.
        self.virialSpeed = np.sqrt(self.G*self.mHalo/self.rDelta)
        
        # Plummer Sphere Parameters
        self.usePlummer = usePlummer
        self.plummerMass = plummerMass # This is given a default value, but it's not used if usePlummer is False
        self.rScalePlummer = rScalePlummer # tracerAttributes[3]. This is given a default value, but it's not used if usePlummer is False
        
        # Creating tracerAttributes is a sloppy way for me to pass in important
        # tracers parameters to methods in nfwFunctions.py, inverseTransforms.py, and nfwFunctions.py
        self.tracerAttributes = np.array((self.rScaleNFW,self.rho0,self.rDelta,self.rScalePlummer))
        
        # Parallelization parameters
        self.nt = numThreads
        self.useParallel = useParallel
        
    def genTracerCoords(self):
        """
        Generates the spatial coordinates of the tracers. Samples from the mass density distribution via inverse transform sampling.
        
        Don't call this function, only call genTracers!
        """
        self.radiiNorm = np.zeros(self.nTracers)
        self.coordAngles = np.zeros((2,self.nTracers))
        
        if self.obsCutoff == False:
            rNormRange = np.array((0.,1.))
        else:
            rNormRange = np.array(((self.earthRadius - self.obsRadius)/self.rDelta,(self.earthRadius + self.obsRadius)/self.rDelta))
            
        if self.useParallel==True:
            if __name__ == '__main__':
                with Pool(processes=self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        ## result_list.append adds the inverseTransform.inverseTransformSamplingRadius method
                        ## to thread k, with arguments specified in the list (len(threadIndices),...etc).
                        result_list.append(
                            pool.apply_async(
                                inverseTransform.inverseTransformSamplingRadius,
                                (len(threadIndices),
                                 self.tracerAttributes,
                                 rNormRange,
                                 self.obsCutoff,
                                 self.obsRadius,
                                 self.earthRadius,
                                 self.usePlummer,
                                )
                            )
                        )
                    ## All the jobs have been assigned to each thread, this loop gets us ready to catch the outputs
                    ## from each thread and store them where they need to go.
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        self.radiiNorm[threadIndices], self.coordAngles[:,threadIndices] = result_list[k].get()
                    ## Closing and joining keeps us from breaking multiprocessing completely.
                    pool.close()
                    pool.join()
        else:
            ## If parallelization isn't an option, this should run it completely in serial with no dependencies on multiprocessing.
            self.radiiNorm, self.coordAngles = inverseTransform.inverseTransformSamplingRadius(self.nTracers, self.tracerAttributes, bracket = rNormRange, obsCutoff = self.obsCutoff, obsRadius = self.obsRadius, earthRadius = self.earthRadius, usePlummer = self.usePlummer)
        # Also create a list of un-normalized radii
        self.radii = self.radiiNorm * self.rDelta
    
    def genPhaseSpaceTable(self):
        """
        Pre-calculates values for the phase space density on a grid of values and builds a spline at each radii along speed.
        
        Don't call this function, only call genTracers!
        """
        
        if self.obsCutoff == False:
            rNormRange = np.array((0.,1.))
        else:
            rNormRange = np.array(((self.earthRadius - self.obsRadius)/self.rDelta,(self.earthRadius + self.obsRadius)/self.rDelta))
            
        self.tabulatedNormRadii = np.linspace(rNormRange[0],rNormRange[1],self.numTabulatedRadii)
        self.tabulatedSpeeds = np.linspace(0,NFW.maxSpeed(.001,self.tracerAttributes),self.numTabulatedSpeeds)
        
        if self.useParallel == True:
            self.psTable = np.zeros((self.numTabulatedRadii,self.numTabulatedSpeeds))
            if __name__ == '__main__':
                with Pool(processes = self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,numTabulatedRadii,self.nt)
                        result_list.append(
                            pool.apply_async(
                                NFW.evalPsdLoop,
                                (
                                    self.tabulatedNormRadii[threadIndices],
                                    self.numTabulatedSpeeds,
                                    self.tabulatedSpeeds,
                                    self.tracerAttributes,
                                )
                            )
                        )
                    for k in range(self.nt):
                        threadIndices = np.arange(k,numTabulatedRadii,self.nt)
                        self.psTable[threadIndices,:] = result_list[k].get()
                    pool.close()
                    pool.join()
        else:
            self.psTable = NFW.evalPsdLoop(self.tabulatedNormRadii,self.numTabulatedSpeeds,self.tabulatedSpeeds,self.tracerAttributes)
        self.psdBivariateSpline = RectBivariateSpline(self.tabulatedNormRadii, self.tabulatedSpeeds, self.psTable, kx = 1, ky = 1)
        
    def genTracerVelocities(self):
        """
        Generates the speeds of the tracer coordinates. Samples from the phase space density via inverse transform sampling.
        
        Don't call this function, only call genTracers!
        """
        self.speeds = np.zeros(self.nTracers)
        self.velocityAngles = np.zeros((2,self.nTracers))
        
        if self.useParallel==True:
            if __name__ == '__main__':
                with Pool(processes=self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        result_list.append(
                            pool.apply_async(
                                inverseTransform.inverseTransformSamplingVelocity,
                                (
                                    len(threadIndices),
                                    self.tracerAttributes,
                                    self.radiiNorm[threadIndices],
                                    self.tabulatedNormRadii,
                                    self.psdBivariateSpline,
                                )
                            )
                        )
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        self.speeds[threadIndices], self.velocityAngles[:,threadIndices] = result_list[k].get()
                    pool.close()
                    pool.join()
        else:
            self.speeds, self.velocityAngles = inverseTransform.inverseTransformSamplingVelocity(self.nTracers, self.tracerAttributes, self.radiiNorm, self.tabulatedNormRadii, self.psdBivariateSpline)
        
    def genTracers(self,numTabulatedRadii,numTabulatedSpeeds, obsCutoff=False, obsRadius=1*10**3, earthRadius=7.91*10**3, debug__ignoreVelocity = False):
        """
        Generates a set of tracer particles with coordinates and velocities that obey an NFW profile. If usePlummer was set to true when the tracerGenerator class was instantiated, the coordinates will obey a plummer distribution while following an NFW velocity profile.
        """
        # Observation region cutoff parameters
        self.obsCutoff = obsCutoff
        self.obsRadius = obsRadius
        self.earthRadius = earthRadius
        
        if (self.earthRadius+self.obsRadius) > self.rDelta:
            print('The maximum observation range, earthRadius + obsRadius = ' + str(self.earthRadius+self.obsRadius) + ' reaches outside the cutoff radius ' + str(self.rDelta) + ', please choose a smaller observation range if you plan to generate observation region-only tracers.')
            return
        if (self.earthRadius-self.obsRadius) < 0:
            print('The maximum observation range, earthRadius - obsRadius = ' + str(self.earthRadius-self.obsRadius) + ' reaches past the galactic center (r=0), which breaks the cutoff procedure and makes it mostly pointless. Please choose a smaller observation range if you plan to generate observation region-only tracers.')
            return
        
        self.numTabulatedRadii = numTabulatedRadii
        self.numTabulatedSpeeds = numTabulatedSpeeds
        
        # Generate spatial coordinates
        print("Generating " + str(self.nTracers) + " coordinates, please wait...")
        time0=time.time()
        self.genTracerCoords()
        time1=time.time()
        print("Coordinates generated in " + str(round(time1-time0,2)) + " seconds.")
        
        # debug__ignoreVelocity is helpful for testing spatial coordinate generation, since velocity generation takes ages
        if debug__ignoreVelocity == False:
            # Pre-tabulate the phase space density function for quick-calls later on
            print("Generating the " +str(self.numTabulatedRadii) + ' by ' + str(self.numTabulatedSpeeds) + ' table of phase space density values')
            time0 = time.time()
            self.genPhaseSpaceTable()
            time1 = time.time()
            print('PSD tabulated in '+ str(round(time1-time0,2)) + " seconds.")

            # Generate velocity coordinates
            print("Generating " + str(self.nTracers) + " velocities, please wait...")
            time0 = time.time()
            self.genTracerVelocities()
            time1 = time.time()
            print("Velocities generated in " + str(round(time1-time0,2)) + " seconds.")

            # Package the coordinates into standard coordinate systems.
            self.sphericalPhaseSpacePoints = [self.radii,
                                              self.coordAngles[0,:],
                                              self.coordAngles[1,:],
                                              self.speeds,
                                              self.velocityAngles[0,:],
                                              self.velocityAngles[1,:]
                                             ]
            self.cartesianPhaseSpacePoints = [self.radii*np.cos(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.sin(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.cos(self.coordAngles[0,:]),
                                              self.speeds*np.cos(self.velocityAngles[1,:])*np.sin(self.velocityAngles[0,:]),
                                              self.speeds*np.sin(self.velocityAngles[1,:])*np.sin(self.velocityAngles[0,:]),
                                              self.speeds*np.cos(self.velocityAngles[0,:])
                                             ]
        else:
            self.sphericalPhaseSpacePoints = [self.radii,
                                              self.coordAngles[0,:],
                                              self.coordAngles[1,:]
                                             ]
            self.cartesianPhaseSpacePoints = [self.radii*np.cos(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.sin(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.cos(self.coordAngles[0,:])
                                             ]
                                  
        print("All " + str(self.nTracers) + " tracers generated.")
        
    def saveTracers(self):
        """
        Saves a copy of the generated tracers.
        """
        timestamp = time.strftime('%b_%d_%Y_%H%M', time.localtime())
        metaLabels = np.array(('nTracers', 'rScaleNFW', 'mwHaloDensity', 'delta',' criticalDensity', 'obsCutoff', 'earthRadius', 'obsRadius', 'usePlummer', 'rScalePlummer', 'plummerMass'))
        tracerAttribs = np.array((self.nTracers, self.rScaleNFW, self.rho0, self.delta, self.rhoCritical, self.obsCutoff, self.earthRadius, self.obsRadius, self.usePlummer, self.rScalePlummer, self.plummerMass))
        phaseSpaceLabels = np.array(('x_pc','y_pc','z_pc','vx_km_s','vy_km_s','vz_km_s'))
        if self.obsCutoff == True:
            if self.usePlummer == True:
                np.savez('generatedTracers//nTracers_' + str(self.nTracers) + '__observRegCutoff_' + str(self.obsCutoff) + '__usesPlummerSpatialDist_' + str(self.usePlummer) + '__time_' + timestamp,metaLabels,tracerAttribs,phaseSpaceLabels,self.cartesianPhaseSpacePoints)
            else:
                np.savez('generatedTracers//nTracers_' + str(self.nTracers) + '__observRegCutoff_' + str(self.obsCutoff) + '__time_' + timestamp,metaLabels,tracerAttribs,phaseSpaceLabels,self.cartesianPhaseSpacePoints)
        else:
            if self.usePlummer == True:
                np.savez('generatedTracers//nTracers_' + str(self.nTracers) + '__usesPlummerSpatialDist_' + str(self.usePlummer) + '__time_' + timestamp,metaLabels,tracerAttribs,phaseSpaceLabels,self.cartesianPhaseSpacePoints)
            else:
                np.savez('generatedTracers//nTracers_' + str(self.nTracers) + '__time_' + timestamp,metaLabels,tracerAttribs,phaseSpaceLabels,self.cartesianPhaseSpacePoints)
```

```python
# General parameters
nTracers = 10**6 # number of tracers to generate

# NFW halo parameters
rScaleNFW = 15*10**3 # parsecs
mwHaloDensity = 3.8 * 10 ** -2 # Solar masses per cubic parsec
delta = 200 # Cutoff scale, NFW halo stops when the density reaches delta*criticalDensity

# Plummer sphere/cluster parameters
flag__usePlummer = True
plummerMass = .5*10**10 # Mass of the plummer sphere (this is the estimated mass of the MW bulge in galaxia's milkywayhalo 2014 fit.)
rScalePlummer = 3*10**3 # scale radius of the plummer sphere (this is the estimated scale radius of the MW bulge in galaxia's milkywayhalo 2014 fit.)
# We're going to cut off generation at whatever rDelta turns out to be, given the NFW halo parameters.

# Observation region cutoff parameters
flag__obsCutoff = False # If set to true, it will only generate samples in a radius obsRadius around the point earthRadius on the x-axis.
obsRadius = 1*10**3 # How far out we want to generate tracers from the position of the earth in parsecs
earthRadius = 7.91*10**3 # Earth <-> Milky Way center distance in parsecs

# Parallelization parameters
flag__useParallel = True # If set to true, intensive generation processes will run in parallel using multiprocessing.
numThreads = 6 # Don't set this higher than the number of available threads.

# Create an instance of the tracerGenerator class
tracers = tracerGenerator(nTracers, rScaleNFW, mwHaloDensity, delta, flag__useParallel, numThreads, flag__usePlummer, plummerMass, rScalePlummer)

# Print out some information about the halo
print('rScaleNFW is : '+str(tracers.rScaleNFW),' parsecs, rDelta is: ' + str(np.round(tracers.rDelta,2)) + ' parsecs.')
print('the NFW halo concentration is: ' + str(np.round(tracers.c,2)) + ', and has a mass of: ' + str(np.round(tracers.mHalo,2)) + ' solar masses.')

# This step generates the 6D coordinates of nTracer # of tracers using the provided characteristics of the halo (and possibly the Plummer sphere as well).
numTabulatedRadii = 400 # 400 or more suggested. Number of radii to sample. Ranges from 0 to rDelta. If flag__obsCutoff==True, it will sample from earthRadius-obsRadius to earthRadius+obsRadius instead.
numTabulatedSpeeds = 400 # 400 or more suggested. Number of speeds to sample. Ranges from 0 to the maximum possible speed in an NFW halo.
tracers.genTracers(numTabulatedRadii, numTabulatedSpeeds, flag__obsCutoff, obsRadius, earthRadius, False)
```

```python
# Saving the generated tracers so we can load them up for density estimation
saveTracers = True
if saveTracers == True:
    tracers.saveTracers()
    
```

```python
# Comparing the generated and theoretical probability distribution function for the tracer radii.
fig, ax1 = plt.subplots(figsize=(10,10))

steps = 200

color = 'tab:blue'
ax1.set_xlabel('Radius (normalized to R200)')
ax1.set_ylabel('Radius PDF')
x_vals, bins, p=ax1.hist(tracers.radiiNorm,steps,color=color,density=True,label="Generated Distribution",align='mid')

color = 'tab:red'
r = np.linspace(0,1,steps)
pdf = []
for i in range(steps):
    pdf.append(NFW.radialPDF(r[i],tracers.tracerAttributes))
ax1.plot(r,pdf, color=color,linewidth=3,label="Exact Distribution")
ax1.legend()
ax1.set_xlim(0,1)

plt.show()
```

<h1>Phase Space Analysis</h1>

```python
nr=1600 # how many radii do we want to sample?
ns=1600 # how many speeds do we want to sample?

if flag__obsCutoff == True:
    maxNormRad = (earthRadius + obsRadius)/tracers.rDelta
    minNormRad = (earthRadius - obsRadius)/tracers.rDelta
else:
    maxNormRad = 1
    minNormRad = 0
    
normRadiusList = np.linspace(minNormRad,maxNormRad,nr)
speedList = np.linspace(0,np.max(tracers.speeds),ns)
radGrid, speedGrid = np.meshgrid(normRadiusList,speedList,sparse=False, indexing='ij')

phaseSpaceGrid = np.zeros((nr,ns))
boxSize = (np.max(tracers.speeds)/ns)*(tracers.rDelta/nr) # has units of (km/s * pc)

updateInt = int(.1*nr)
for i in range(nr):
    if (i+1)%updateInt == 0:
        print("Calculation is "+str(100*(i+1)/nr)+" % done")
    for j in range(ns):
        phaseSpaceGrid[i,j] = boxSize*tracers.rDelta**2*(4*np.pi*radGrid[i,j]*speedGrid[i,j])**2*tracers.psdBivariateSpline(radGrid[i,j],speedGrid[i,j])

phaseSpaceIntApprox = np.sum(phaseSpaceGrid)
print("Integrated phase space reaches: " + str(phaseSpaceIntApprox))
```

```python
fig,(ax1, ax2)= plt.subplots(1,2,figsize=(32,18))

ax1.imshow(np.transpose(phaseSpaceGrid),origin='lower',aspect='auto',cmap='RdBu_r')
ax1.set_ylabel('Speed (km/s)')
ax1.set_xlabel('Radius (pc)')
num_x_labels = 12
step_r = int(nr/(num_x_labels-1))
ax1.set_xticks(np.arange(0,nr,step_r))
ax1.set_xticklabels(np.round(tracers.rDelta*normRadiusList[::step_r],1))
num_y_labels = 10
step_s = int(ns/(num_y_labels-1))
ax1.set_yticks(np.arange(0,ns,step_s))
ax1.set_yticklabels(np.round(speedList[::step_s],0))
ax1.set_title('Theory')

ax2.hexbin(
    tracers.radii, 
    tracers.speeds,
    mincnt=1,gridsize=(30,30),cmap='RdBu_r'
)
x=tracers.rDelta*normRadiusList
y=speedList
ax2.contour(x,y,np.transpose(phaseSpaceGrid),cmap='RdBu_r',linewidths=3, levels=np.linspace(np.min(phaseSpaceGrid[np.nonzero(phaseSpaceGrid)]),np.max(phaseSpaceGrid),10))
ax2.set_xlabel("Radius (pc)")
ax2.set_ylabel("Speed (km/s)")
ax2.set_title("Generated (w/ theory overlaid as a contour)")
```
