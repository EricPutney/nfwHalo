```python
import numpy as np
import scipy
from scipy import integrate
from numpy.random import default_rng
from multiprocessing import Pool
import nfwFunctions as NFW
import inverseTransform
import time
import sys

from scipy import special
from scipy.interpolate import interp1d
import matplotlib as mpl
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
class nfwHalo(): # Our NFW halo object
    def __init__(self, nTracers, rScale, rho0, delta, rhoCritical = 1.26*10**-7, useParallel = True, numThreads = 1, obsCutoff=False, obsRadius=1*10**3, earthRadius=7.91*10**3):
        """
        NFW Halo. This init just stores some of the arguments passed in and computes some important/handy quantities.
        
        Important properties of a nfwHalo object:
            nTracers - number of tracer stars.
            rScale - often called "a" in the literature, this is the scale radius for the NFW profile.
            rho0 - scale density for NFW profile.
            delta - cutoff scale, usually 200.
            c - the concentration i.e. ratio between rDelta and rScale.
            rDelta - the radius at which we cut off tracer generation.
            mHalo - mass of the halo.
            virialSpeed - virial speed of the halo.
        
        Important tracer properties:
            radii - radii of generated tracers (run genTracers first)
            normRadii - radii of generated tracers normalized to the cutoff radius rDelta (run genTracers first)
            speeds - speeds of the generated tracers (run genTracers first)
            coordAngles - the spherical coordinate angles (theta, phi) for each tracer in position space (run genTracers first)
            velocityAngles - the spherical coordinate angles (theta, phi) for each tracer in velocity space (run genTracers first)
            sphericalPhaseSpacePoints - contains the 6D phase space of all tracers in spherical coordinates
            cartesianPhaseSpacePoints - contains the 6D phase space of all tracers in cartesian coordinates
        """
        self.nTracers = nTracers # Halo attribute 0
        self.rScale = rScale # Halo attribute 1
        self.rho0 = rho0 # Halo attribute 2
        self.delta = delta # Halo attribute 3
        self.rhoCritical = rhoCritical # Halo attribute 4
        rhoNorm = self.rho0/(self.delta*self.rhoCritical)
        cShortcut = ((3.*np.sqrt(81.*rhoNorm**2. + 12.*rhoNorm) + 27.*rhoNorm + 2.)/2.)**(1./3.)
        self.c = ( cShortcut + 1./cShortcut - 2 )/3. # Halo attribute 5
        self.rDelta = self.c*self.rScale # Halo attribute 6
        self.G = 4.3009*10**-3 # Newton's constant, (pc/solar mass) * (km/second)^2. Halo attribute 7
        self.mHalo = (4*np.pi*self.rho0*self.rScale**3)*NFW.gFcn(self.c) # Should be in units of solar masses. Halo attribute 8
        self.virialSpeed = np.sqrt(self.G*self.mHalo/self.rDelta) # Halo attribute 9
        
        # Creating haloAttributes is a very sloppy way for me to pass in important
        # halo parameters to methods in nfwFunctions.py
        self.haloAttributes = np.array((self.nTracers,self.rScale,self.rho0,self.delta,self.rhoCritical,self.c,self.rDelta,self.G,self.mHalo,self.virialSpeed))
        
        # Parallelization parameters
        self.nt = numThreads
        self.useParallel = useParallel
        
        # Observation point cutoff parameters
        self.obsCutoff = obsCutoff
        self.obsRadius = obsRadius
        self.earthRadius = earthRadius
        
        if (self.earthRadius+self.obsRadius) > self.rDelta:
            print('The maximum observation range, earthRadius + obsRadius = ' + str(self.earthRadius+self.obsRadius) + ' reaches outside the cutoff radius ' + str(self.rDelta) + ', please choose a smaller observation range if you plan to generate an observation region-only halo.')
            return
        if (self.earthRadius-self.obsRadius) < 0:
            print('The maximum observation range, earthRadius - obsRadius = ' + str(self.earthRadius-self.obsRadius) + ' reaches past the galactic center (r=0), which breaks the cutoff procedure and makes it mostly pointless. Please choose a smaller observation range if you plan to generate an observation region-only halo.')
            return
        
    def genTracerCoords(self):
        """
        Builds an array of tracer speeds. Samples from the mass density distribution via inverse transform sampling.
        """
        self.radiiNorm = np.zeros(self.nTracers)
        self.coordAngles = np.zeros((2,self.nTracers))
        
        if self.obsCutoff == False:
            rNormRange = np.array((0.,1.))
        else:
            rNormRange = np.array(((self.earthRadius + self.obsRadius)/self.rDelta,(self.earthRadius - self.obsRadius)/self.rDelta))
            
        if self.useParallel==True:
            if __name__ == '__main__':
                with Pool(processes=self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        ## result_list.append adds the inverseTransform.inverseTransformSamplingRadius method
                        ## to thread k, with arguments specified in the list (len(threadIndices),...etc).
                        result_list.append(
                            pool.apply_async(
                                inverseTransform.inverseTransformSamplingRadius,
                                (len(threadIndices),
                                 self.haloAttributes,
                                 rNormRange,
                                 self.obsCutoff,
                                 self.obsRadius,
                                 self.earthRadius,
                                )
                            )
                        )
                    ## All the jobs have been assigned to each thread, this loop gets us ready to catch the outputs
                    ## from each thread and store them where they need to go.
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        self.radiiNorm[threadIndices], self.coordAngles[:,threadIndices] = result_list[k].get()
                    ## Closing and joining keeps us from breaking multiprocessing completely.
                    pool.close()
                    pool.join()
        else:
            ## If parallelization isn't an option, this should run it completely in serial with no dependencies on multiprocessing.
            self.radiiNorm, self.coordAngles = inverseTransform.inverseTransformSamplingRadius(self.nTracers, self.haloAttributes, bracket = rNormRange, obsCutoff = self.obsCutoff, obsRadius = self.obsRadius, earthRadius = self.earthRadius)
        # Also create a list of un-normalized radii
        self.radii = self.radiiNorm * self.rDelta
    
    def genPhaseSpaceTable(self):
        """
        Pre-calculates values for the phase space density on a grid of values and builds a spline at each radii along speed.
        """
        
        if self.obsCutoff == False:
            rNormRange = np.array((0.,1.))
        else:
            rNormRange = np.array(((self.earthRadius + self.obsRadius)/self.rDelta,(self.earthRadius - self.obsRadius)/self.rDelta))
            
        self.tabulatedNormRadii = np.linspace(rNormRange[0],rNormRange[1],self.numTabulatedRadii)
        self.tabulatedSpeeds = np.linspace(0,NFW.maxSpeed(.001,self.haloAttributes),self.numTabulatedSpeeds)
        
        if self.useParallel == True:
            self.psTable = np.zeros((self.numTabulatedRadii,self.numTabulatedSpeeds))
            self.speedSplines = [None] * self.numTabulatedRadii
            if __name__ == '__main__':
                with Pool(processes = self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,numTabulatedRadii,self.nt)
                        result_list.append(
                            pool.apply_async(
                                NFW.evalPsdLoop,
                                (
                                    self.tabulatedNormRadii[threadIndices],
                                    self.numTabulatedSpeeds,
                                    self.tabulatedSpeeds,
                                    self.haloAttributes,
                                )
                            )
                        )
                    for k in range(self.nt):
                        threadIndices = np.arange(k,numTabulatedRadii,self.nt)
                        splineOut, psTableOut = result_list[k].get()
                        for q in range(len(threadIndices)):
                            self.speedSplines[threadIndices[q]] = splineOut[q]
                        self.psTable[threadIndices,:] = psTableOut
                    pool.close()
                    pool.join()
        else:
            self.speedSplines, self.psTable = NFW.evalPsdLoop(self.tabulatedNormRadii,self.numTabulatedSpeeds,self.tabulatedSpeeds,self.haloAttributes)
                

                        
    def genTracerVelocities(self):
        """
        Builds a list of tracer speeds. Samples from the phase space density via inverse transform sampling.
        """
        self.speeds = np.zeros(self.nTracers)
        self.velocityAngles = np.zeros((2,self.nTracers))
        
        if self.useParallel==True:
            if __name__ == '__main__':
                with Pool(processes=self.nt) as pool:
                    result_list = []
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        result_list.append(
                            pool.apply_async(
                                inverseTransform.inverseTransformSamplingVelocity,
                                (
                                    len(threadIndices),
                                    self.haloAttributes,
                                    self.radiiNorm[threadIndices],
                                    self.tabulatedNormRadii,
                                    self.speedSplines,
                                )
                            )
                        )
                    for k in range(self.nt):
                        threadIndices = np.arange(k,self.nTracers,self.nt)
                        self.speeds[threadIndices], self.velocityAngles[:,threadIndices] = result_list[k].get()
                    pool.close()
                    pool.join()
        else:
            self.speeds, self.velocityAngles = np.array(inverseTransform.inverseTransformSamplingVelocity(self.nTracers, self.haloAttributes, self.radiiNorm, self.psTable, self.tabulatedNormRadii, self.speedSplines,))
        
    def genTracers(self,numTabulatedRadii,numTabulatedSpeeds, debug__ignoreVelocity = False):
        """
        Automatically generates a set of tracer particles with coordinates and velocities that obey an NFW profile.
        """
        self.numTabulatedRadii = numTabulatedRadii
        self.numTabulatedSpeeds = numTabulatedSpeeds
        
        # Generate spatial coordinates
        print("Generating " + str(self.nTracers) + " coordinates, please wait...")
        time0=time.time()
        self.genTracerCoords()
        time1=time.time()
        print("Coordinates generated in " + str(round(time1-time0,2)) + " seconds.")
        
        if debug__ignoreVelocity == False:
            # Pre-tabulate the phase space density function for quick-calls later on
            print("Generating the " +str(self.numTabulatedRadii) + ' by ' + str(self.numTabulatedSpeeds) + ' table of phase space density values')
            time0 = time.time()
            self.genPhaseSpaceTable()
            time1 = time.time()
            print('PSD tabulated in '+ str(round(time1-time0,2)) + " seconds.")

            # Generate velocity coordinates
            print("Generating " + str(self.nTracers) + " velocities, please wait...")
            time0 = time.time()
            self.genTracerVelocities()
            time1 = time.time()
            print("Velocities generated in " + str(round(time1-time0,2)) + " seconds.")

            # Package the coordinates into standard coordinate systems.
            self.sphericalPhaseSpacePoints = [self.radii,
                                              self.coordAngles[0,:],
                                              self.coordAngles[1,:],
                                              self.speeds,
                                              self.velocityAngles[0,:],
                                              self.velocityAngles[1,:]
                                             ]
            self.cartesianPhaseSpacePoints = [self.radii*np.cos(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.sin(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.cos(self.coordAngles[0,:]),
                                              self.speeds*np.cos(self.velocityAngles[1,:])*np.sin(self.velocityAngles[0,:]),
                                              self.speeds*np.sin(self.velocityAngles[1,:])*np.sin(self.velocityAngles[0,:]),
                                              self.speeds*np.cos(self.velocityAngles[0,:])
                                             ]
        else:
            self.sphericalPhaseSpacePoints = [self.radii,
                                              self.coordAngles[0,:],
                                              self.coordAngles[1,:]
                                             ]
            self.cartesianPhaseSpacePoints = [self.radii*np.cos(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.sin(self.coordAngles[1,:])*np.sin(self.coordAngles[0,:]),
                                              self.radii*np.cos(self.coordAngles[0,:])
                                             ]
                                  
        print("All " + str(self.nTracers) + " tracers generated.")
        
    def saveHalo(self):
        """
        Saves a copy of the tracer information.
        """
        t = time.localtime()
        timestamp = time.strftime('%b_%d_%Y_%H%M', t)
        metaLabels = np.array(('nTracers','rScale','mwHaloDensity','delta','criticalDensity'))
        haloAttrib = np.array((self.nTracers,self.rScale,self.rho0,self.delta,self.rhoCritical))
        phaseSpaceLabels = np.array(('x_pc','y_pc','z_pc','vx_km_s','vy_km_s','vz_km_s'))
        np.savez('generatedHalos//nfwGeneratedTracers_'+timestamp,metaLabels,haloAttrib,phaseSpaceLabels,self.cartesianPhaseSpacePoints)
```

```python
# Let's make a halo!

# NFW halo parameters
nTracers = 10**5 # number of tracers to generate
rScale = 15*10**3 # parsecs
mwHaloDensity = 3.8 * 10 ** -2 # Solar masses per cubic parsec
delta = 200
criticalDensity = 1.26 * 10 ** -7 # Solar masses per cubic parsec

# Observation point cutoff parameters
obsCutoff = True
obsRadius = 5*10**3
earthRadius = 7.91*10**3

# Parallelization parameters
useParallel = True
numThreads = 6

# Create an instance of the nfwHalo class
mwHalo = nfwHalo(nTracers,rScale,mwHaloDensity,delta,criticalDensity,useParallel,numThreads,obsCutoff,obsRadius,earthRadius)

# Print out some information about the halo
print('rScale is : '+str(mwHalo.rScale),', rDelta is: ' + str(mwHalo.rDelta))
print('the concentration must then be: ' + str(mwHalo.c))

# This step generates the 6D coordinates of nTracer # of tracers using the provided characteristics of the halo.
numTabulatedRadii = 200
numTabulatedSpeeds = 200
mwHalo.genTracers(numTabulatedRadii,numTabulatedSpeeds)
```

```python
# Saving the generated halo so we can load it up in another file
saveHalo = True
if saveHalo == True:
    mwHalo.saveHalo()
    
```

```python
# Comparing the generated and theoretical probability distribution function for the halo radii.
fig, ax1 = plt.subplots(figsize=(10,10))

steps = 200

color = 'tab:blue'
ax1.set_xlabel('Radius (normalized to R200)')
ax1.set_ylabel('Radius PDF')
x_vals, bins, p=ax1.hist(mwHalo.radiiNorm,steps,color=color,density=True,label="Generated Distribution",align='mid')

color = 'tab:red'
r = np.linspace(0,1,steps)
pdf = []
for i in range(steps):
    pdf.append(NFW.radialPDF(r[i],mwHalo.haloAttributes))
ax1.plot(r,pdf, color=color,linewidth=3,label="Exact Distribution")
ax1.legend()
ax1.set_xlim(0,1)

plt.show()
```

<h1>Phase Space Analysis</h1>

```python
nr=400 # how many radii do we want to sample?
ns=400 # how many speeds do we want to sample?

if obsCutoff == True:
    maxNormRad = (earthRadius + obsRadius)/mwHalo.rDelta
    minNormRad = (earthRadius - obsRadius)/mwHalo.rDelta
else:
    maxNormRad = 1
    minNormRad = 0
    
normRadiusList = np.linspace(minNormRad,maxNormRad,nr)
speedList = np.linspace(0,np.max(mwHalo.speeds),ns)
radGrid, speedGrid = np.meshgrid(normRadiusList,speedList,sparse=False, indexing='ij')

phaseSpaceGrid = np.zeros((nr,ns))
boxSize = (np.max(mwHalo.speeds)/ns)*(mwHalo.rDelta/nr) # has units of (km/s * pc)

updateInt = int(.1*nr)
for i in range(nr):
    if (i+1)%updateInt == 0:
        print("Calculation is "+str(100*(i+1)/nr)+" % done")
    for j in range(ns):
        phaseSpaceGrid[i,j] = boxSize*mwHalo.rDelta**2*(4*np.pi*radGrid[i,j]*speedGrid[i,j])**2*NFW.phaseSpaceDistributionFromTable(radGrid[i,j],speedGrid[i,j],mwHalo.tabulatedNormRadii,mwHalo.speedSplines,mwHalo.haloAttributes)

phaseSpaceIntApprox = np.sum(phaseSpaceGrid)
print("Integrated phase space reaches: " + str(phaseSpaceIntApprox))
```

```python
fig,(ax1, ax2)= plt.subplots(1,2,figsize=(16,9))

ax1.imshow(np.transpose(phaseSpaceGrid),origin='lower',aspect='auto',cmap='RdBu_r')
ax1.set_ylabel('Speed (km/s)')
ax1.set_xlabel('Radius (pc)')
num_x_labels = 6
step_r = int(nr/(num_x_labels-1))
ax1.set_xticks(np.arange(0,nr,step_r))
ax1.set_xticklabels(np.round(mwHalo.rDelta*normRadiusList[::step_r],1))
num_y_labels = 10
step_s = int(ns/(num_y_labels-1))
ax1.set_yticks(np.arange(0,ns,step_s))
ax1.set_yticklabels(np.round(speedList[::step_s],0))
ax1.set_title('Theory')

ax2.hexbin(
    mwHalo.radii, 
    mwHalo.speeds,
    mincnt=1,gridsize=(30,30),cmap='RdBu_r'
)
x=mwHalo.rDelta*normRadiusList
y=speedList
ax2.contour(x,y,np.transpose(phaseSpaceGrid),cmap='RdBu_r',linewidths=3, levels=np.linspace(np.min(phaseSpaceGrid[np.nonzero(phaseSpaceGrid)]),np.max(phaseSpaceGrid),10))
ax2.set_xlabel("Radius (pc)")
ax2.set_ylabel("Speed (km/s)")
ax2.set_title("Generated (w/ theory overlaid as a contour)")
```
