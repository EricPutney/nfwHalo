```python
import numpy as np
import scipy
from scipy import integrate
from numpy.random import default_rng
from multiprocessing import Pool
import inverseTransform
import time

from scipy import special
import matplotlib as mpl
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import clear_output

plt.rcParams['figure.autolayout'] = False
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 14
mpl.rc('axes',edgecolor='k')
plt.rcParams['xtick.color'] = 'k'
plt.rcParams['ytick.color'] = 'k'
```

```python
class halo():
    def __init__(self, rScale, rho0, delta, rhoCritical = 1.26*10**-7):
        self.rho0 = rho0
        self.rScale = rScale
        self.G = 4.3009*10**-3 # (pc/solar mass) * (km/s)^2
        self.rhoCritical = rhoCritical
        self.delta = delta
        rhoNorm = rho0/(delta*rhoCritical)
        cTerm = ((3.*np.sqrt(81.*rhoNorm**2. + 12.*rhoNorm) + 27.*rhoNorm + 2.)/2.)**(1./3.)
        self.c = ( cTerm + 1./cTerm - 2 )/3.
        self.rDelta = self.c*self.rScale
        self.haloMass = (4.*np.pi*self.rho0*self.rScale**3)*self.gFcn(self.c)
        self.virialSpeed = np.sqrt(self.G*self.haloMass/self.rDelta)
        self.energyNorm = 4.*np.pi*self.rho0*self.G*self.rScale**2
        
    def relativeEnergy(self,r,s,usePhi0=True):
        radiusNorm = r/self.rDelta
        if usePhi0 == True:
            Phi0 = -self.energyNorm*np.log(1+self.c)/self.c
        else:
            Phi0 = 0
        Phi = -self.energyNorm*np.log(1+self.c*radiusNorm)/(self.c*radiusNorm)
        return -Phi + Phi0 - (1/2)*s**2
    
    def relativePotential(self,r,usePhi0=True):
        radiusNorm = r/self.rDelta
        if usePhi0 == True:
            Phi0 = -self.energyNorm*np.log(1+self.c)/self.c
        else:
            Phi0 = 0
        Phi = -self.energyNorm*np.log(1+self.c*radiusNorm)/(self.c*radiusNorm)
        return -Phi + Phi0
    
    def gFcn(self,x):
        return np.log(1+x)-(x/(1+x))
    
    def maxSpeed(self,r,usePhi0=True):
        return np.sqrt(2*self.relativePotential(r,usePhi0))
        
    def PSD_MB(self,r,s,usePhi0=True):
        eps = self.relativeEnergy(r,s,usePhi0)
        if eps<0:
            return 0
        else:
            radiusNorm = r/self.rDelta
            disp = np.sqrt(self.virialSpeed**2*\
                        ((self.c**2*radiusNorm*(1+self.c*radiusNorm)**2)/self.gFcn(self.c))*\
                        integrate.quad(lambda y: self.gFcn(y)/(y**3*(1+y)**2), self.c*radiusNorm, np.inf)[0]
                        )
            return (1/(self.c*radiusNorm*(1+self.c*radiusNorm)**2)) * np.exp(-s**2/(2*disp**2))/( np.sqrt(128*np.pi**5) * self.rScale**3 * disp**3 * self.gFcn(self.c))
        
    def PSD_Eddington(self,r,s,usePhi0=True):
        eps = self.relativeEnergy(r,s,usePhi0)
        if eps<0:
            return 0
        else:
            epsNorm = eps/self.energyNorm
            if usePhi0 == True:
                PhiNorm = -np.log(1+self.c)/self.c
            else:
                PhiNorm = 0
            prefactor = 1./(np.sqrt(128)*np.pi**3*self.rScale**3*self.energyNorm**(3./2.)*self.gFcn(self.c))

            integrand = lambda PsiNorm: -(PsiNorm-PhiNorm)**3*(1+(self.A_Norm(PsiNorm,usePhi0)/(PsiNorm-PhiNorm)))*(2+(1+(self.A_Norm(PsiNorm,usePhi0)/(1+self.A_Norm(PsiNorm,usePhi0))))*((3*self.A_Norm(PsiNorm,usePhi0)/(PsiNorm-PhiNorm))+2))/(np.sqrt(np.abs(PsiNorm-epsNorm))*self.A_Norm(PsiNorm,usePhi0)**2*(1+self.A_Norm(PsiNorm,usePhi0))**2*(PsiNorm-PhiNorm+self.A_Norm(PsiNorm,usePhi0)))
            return prefactor*integrate.quad(integrand,0,epsNorm)[0]
    
    def A_Norm(self,PsiNorm,usePhi0=True):
        if usePhi0 == True:
            PhiNorm = -np.log(1+self.c)/self.c
        else:
            PhiNorm = 0
        return np.real(special.lambertw(-(PsiNorm-PhiNorm)*np.exp(-(PsiNorm-PhiNorm)),-1))
    
    def mass_density(self,r):
        return self.rho0/((r/self.rScale)*(1+r/self.rScale)**2)
            
    #def A(self,psi):
    #    b = 4*np.pi*self.rho0*self.G*self.rScale**2
    #    gamma = np.log(1+self.c)/self.c
    #    return np.real(special.lambertw(-(psi/b+gamma)*np.exp(-(psi/b+gamma)),-1))
            
    #def nu(self,psi):
    #    b = 4*np.pi*self.rho0*self.G*self.rScale**2
    #    gamma = np.log(1+self.c)/self.c
    #    return (-1/(4*np.pi*self.rScale**3*self.gFcn(self.c)))*(psi/b+gamma)**3/(self.A(psi)**2*((psi/b+gamma)+self.A(psi)))
    
    #def dNu_dPsi_first(self,psi):
    #    b = 4*np.pi*self.rho0*self.G*self.rScale**2
    #    gamma = np.log(1+self.c)/self.c
    #    return (self.nu(psi)/(b*(1+self.A(psi))))*((3*self.A(psi)/(psi/b+gamma))+2)
        
    #def dNu_dPsi_second(self,psi):
    #    b = 4*np.pi*self.rho0*self.G*self.rScale**2
    #    gamma = np.log(1+self.c)/self.c
    #    return (self.nu(psi)/(b*(1+self.A(psi)))**2)*(1+(self.A(psi))/(psi/b+gamma))*(2+(1+(self.A(psi)/(1+self.A(psi))))*((3*self.A(psi)/(psi/b+gamma))+2))
    
```

```python
rScale = 10*10**3 # Scale radius of 10 kpc
mwHaloDensity = 4.88664 * 10 ** -2 # Solar masses per cubic parsec. This is the density that gives a mass of 10^12 for rScale = 10,000 kPc
delta = 200 # cutoff scale
criticalDensity = 1.26 * 10 ** -7 # Solar masses per cubic parsec

# Create an instance of the nfwHalo class
mwHalo = halo(rScale,mwHaloDensity,delta,criticalDensity)
print("This halo has a mass of " + "{:e}".format(mwHalo.haloMass) + " solar masses, and a concentration of " + str(mwHalo.c))
```

```python
nRadii = 200
nSpeeds = 200

maxRadius = mwHalo.rDelta
testradii = np.linspace(0.001,maxRadius,100)
ms = np.zeros(100)
for k in range(100):
    ms[k] = mwHalo.maxSpeed(testradii[k],usePhi0=False)
maxSpeed = np.max(ms)

radii = np.linspace(5,maxRadius,nRadii) # in parsecs
speeds = np.linspace(.1,maxSpeed,nSpeeds) # in km/s
```

```python
pdfEddingtonImage = np.zeros((nRadii,nSpeeds))
pdfEddingtonNoPhi0Image = np.zeros((nRadii,nSpeeds))
pdfMBImage = np.zeros((nRadii,nSpeeds))
pdfMBNoPhi0Image = np.zeros((nRadii,nSpeeds))
relativeEnergies = np.zeros((nRadii,nSpeeds))
relativeEnergies_noPhi0 = np.zeros((nRadii,nSpeeds))

for i in range(nRadii):
    clear_output(wait=True)
    print(i/nRadii)
    for j in range(nSpeeds):
        pdfEddingtonImage[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_Eddington(radii[i],speeds[j])
        pdfEddingtonNoPhi0Image[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_Eddington(radii[i],speeds[j],usePhi0=False)
        pdfMBImage[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_MB(radii[i],speeds[j])
        pdfMBNoPhi0Image[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_MB(radii[i],speeds[j],usePhi0=False)
        relativeEnergies[i,j] = mwHalo.relativeEnergy(radii[i],speeds[j])
        relativeEnergies_noPhi0[i,j] = mwHalo.relativeEnergy(radii[i],speeds[j],usePhi0=False)
        
```

```python
fig, axs = plt.subplots(2,2,figsize=(20,20))
axis1 = axs[0,0].imshow(np.transpose(pdfEddingtonImage),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
axs[0,0].set_xticks(np.arange(0,nRadii,step_r))
axs[0,0].set_xticklabels(np.round(radii[::step_r]))
axs[0,0].set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
axs[0,0].set_yticks(np.arange(0,nSpeeds,step_s))
axs[0,0].set_yticklabels(np.round(speeds[::step_s]))
axs[0,0].set_ylabel('Speed (km/s)')
axs[0,0].set_title('Eddington NFW Phase Space')
fig.colorbar(axis1,ax=axs[0,0])

axs[0,0].contour(np.transpose(relativeEnergies),levels=[0],linewidths=3)

axis2 = axs[0,1].imshow(np.transpose(pdfEddingtonNoPhi0Image),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
axs[0,1].set_xticks(np.arange(0,nRadii,step_r))
axs[0,1].set_xticklabels(np.round(radii[::step_r]))
axs[0,1].set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
axs[0,1].set_yticks(np.arange(0,nSpeeds,step_s))
axs[0,1].set_yticklabels(np.round(speeds[::step_s]))
axs[0,1].set_ylabel('Speed (km/s)')
axs[0,1].set_title('Eddington NFW Phase Space with Phi_0 = 0')
fig.colorbar(axis2,ax=axs[0,1])

axs[0,1].contour(np.transpose(relativeEnergies_noPhi0),levels=[0],linewidths=3)

axis3 = axs[1,0].imshow(np.transpose(pdfMBImage),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
axs[1,0].set_xticks(np.arange(0,nRadii,step_r))
axs[1,0].set_xticklabels(np.round(radii[::step_r]))
axs[1,0].set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
axs[1,0].set_yticks(np.arange(0,nSpeeds,step_s))
axs[1,0].set_yticklabels(np.round(speeds[::step_s]))
axs[1,0].set_title('Maxwell-Boltzmann NFW Phase Space')
axs[1,0].set_ylabel('Speed (km/s)')
fig.colorbar(axis3,ax=axs[1,0])

axs[1,0].contour(np.transpose(relativeEnergies),levels=[0],linewidths=3)

axis4 = axs[1,1].imshow(np.transpose(pdfMBNoPhi0Image),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
axs[1,1].set_xticks(np.arange(0,nRadii,step_r))
axs[1,1].set_xticklabels(np.round(radii[::step_r]))
axs[1,1].set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
axs[1,1].set_yticks(np.arange(0,nSpeeds,step_s))
axs[1,1].set_yticklabels(np.round(speeds[::step_s]))
axs[1,1].set_title('Maxwell-Boltzmann NFW Phase Space with Phi_0 = 0')
axs[1,1].set_ylabel('Speed (km/s)')
fig.colorbar(axis3,ax=axs[1,1])

axs[1,1].contour(np.transpose(relativeEnergies_noPhi0),levels=[0],linewidths=3)
```

```python
nRadii = 100
maxRadius = mwHalo.rDelta

radii = np.linspace(5,maxRadius,nRadii) # in parsecs

massDensity_Eddington = np.zeros(nRadii)
massDensity_Eddington_noPhi0 = np.zeros(nRadii)
massDensity = np.zeros(nRadii)

for i in range(nRadii):
    clear_output(wait=True)
    print(i/nRadii)
    massDensity_Eddington[i] = mwHalo.haloMass*integrate.quad(lambda x: (4*np.pi*x**2)*mwHalo.PSD_Eddington(radii[i],x),0,mwHalo.maxSpeed(radii[i]))[0]
    massDensity_Eddington_noPhi0[i] = mwHalo.haloMass*integrate.quad(lambda x: (4*np.pi*x**2)*mwHalo.PSD_Eddington(radii[i],x,usePhi0=False),0,mwHalo.maxSpeed(radii[i],usePhi0=False))[0]
    massDensity[i] = mwHalo.mass_density(radii[i])
```

```python
plt.figure()
plt.plot(radii,np.log10(massDensity_Eddington),label='Mass Density from Eddington')
plt.plot(radii,np.log10(massDensity_Eddington_noPhi0),label='Mass Density from Eddington, Phi0 = 0',linewidth=5)
plt.plot(radii,np.log10(massDensity),label='Mass Density')
plt.xlabel('Radius [pc]')
plt.ylabel('log10(mass density) [Solar mass/pc^3]')
plt.legend()
```
