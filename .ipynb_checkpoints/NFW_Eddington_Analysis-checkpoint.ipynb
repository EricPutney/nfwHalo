```python
import numpy as np
import scipy
from scipy import integrate
from numpy.random import default_rng
from multiprocessing import Pool
import inverseTransform
import time

from scipy import special
import matplotlib as mpl
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import clear_output

plt.rcParams['figure.autolayout'] = False
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 14
mpl.rc('axes',edgecolor='k')
plt.rcParams['xtick.color'] = 'k'
plt.rcParams['ytick.color'] = 'k'
```

```python
class halo():
    def __init__(self, rScale, rho0, delta, rhoCritical = 1.26*10**-7):
        self.rho0 = rho0
        self.rScale = rScale
        self.G = 4.3009*10**-3 # (pc/solar mass) * (km/s)^2
        self.rhoCritical = rhoCritical
        self.delta = delta
        rhoNorm = rho0/(delta*rhoCritical)
        cTerm = ((3.*np.sqrt(81.*rhoNorm**2. + 12.*rhoNorm) + 27.*rhoNorm + 2.)/2.)**(1./3.)
        self.c = ( cTerm + 1./cTerm - 2 )/3.
        self.rDelta = self.c*self.rScale
        self.haloMass = (4.*np.pi*self.rho0*self.rScale**3)*self.gFcn(self.c)
        self.virialSpeed = np.sqrt(self.G*self.haloMass/self.rDelta)
        self.energyNorm = 4.*np.pi*self.rho0*self.G*self.rScale**2
        
    def relativeEnergy(self,r,s):
        radiusNorm = r/self.rDelta
        Phi = -self.energyNorm*np.log(1+self.c*radiusNorm)/(self.c*radiusNorm)
        return -Phi - (1/2)*s**2
    
    def relativePotential(self,r):
        radiusNorm = r/self.rDelta
        Phi = -self.energyNorm*np.log(1+self.c*radiusNorm)/(self.c*radiusNorm)
        return -Phi
    
    def gFcn(self,x):
        return np.log(1+x)-(x/(1+x))
    
    def maxSpeed(self,r):
        return np.sqrt(2*self.relativePotential(r))
        
    def PSD_MB(self,r,s):
        eps = self.relativeEnergy(r,s)
        if eps<0:
            return 0
        else:
            radiusNorm = r/self.rDelta
            disp = np.sqrt(self.virialSpeed**2*\
                        ((self.c**2*radiusNorm*(1+self.c*radiusNorm)**2)/self.gFcn(self.c))*\
                        integrate.quad(lambda y: self.gFcn(y)/(y**3*(1+y)**2), self.c*radiusNorm, np.inf)[0]
                        )
            return (1/(self.c*radiusNorm*(1+self.c*radiusNorm)**2)) * np.exp(-s**2/(2*disp**2))/( np.sqrt(128*np.pi**5) * self.rScale**3 * disp**3 * self.gFcn(self.c))
        
    def PSD_Eddington(self,r,s):
        EpsNorm = self.relativeEnergy(r,s)/self.energyNorm
        if (EpsNorm<0 or np.abs(EpsNorm - 1) < 1e-3 or np.isnan(EpsNorm)==True):
            # This is a shallow attempt to control issues when EpsNorm is close to 1. This will not always work since special functions are a curse.
            return 0
        else:
            prefactor = 1./(np.sqrt(128)*np.pi**3*self.rScale**3*self.energyNorm**(3./2.)*self.gFcn(self.c))
            integrand = lambda PsiNorm: -(PsiNorm)**3*(1+(self.A(PsiNorm)/(PsiNorm)))*(2+(1+(self.A(PsiNorm)/(1+self.A(PsiNorm))))*((3*self.A(PsiNorm)/(PsiNorm))+2))/(np.sqrt(np.abs(PsiNorm-EpsNorm))*self.A(PsiNorm)**2*(1+self.A(PsiNorm))**2*(PsiNorm+self.A(PsiNorm)))
            return prefactor*integrate.quad(integrand,0,EpsNorm,epsrel=1e-12)[0]
    
    def A(self,PsiNorm):
            return np.real(special.lambertw(-PsiNorm*np.exp(-PsiNorm),k=-1))
    
    def mass_density(self,r):
        return self.rho0/((r/self.rScale)*(1+r/self.rScale)**2)
    
    # The code below was my attempt to write PSD_Eddington in a simplified way, but something is very broken. The PSD_Eddington function above works properly, though it is a little ugly.
    #def nu(self,PsiNorm):
    #    nu_0 = (4*np.pi*self.rScale**3*self.gFcn(self.c))**(-1)
    #    return -nu_0*(PsiNorm**3)/(self.A(PsiNorm)**2*(PsiNorm + self.A(PsiNorm)))
    #
    #def nu_2nd_derivative(self,PsiNorm):
    #    return self.nu(PsiNorm)*((1+(self.A(PsiNorm)/PsiNorm))/((1 + self.A(PsiNorm))**3))*( 3*(self.A(PsiNorm)/PsiNorm)*(2*self.A(PsiNorm) + 1) + 2*(3*self.A(PsiNorm) + 2) )
    #    
    #def nu_2nd_derivative_2(self,PsiNorm):
    #    return (self.nu(PsiNorm)/((1+self.A(PsiNorm))**2))*(1+self.A(PsiNorm)/PsiNorm)*(2+(1+(self.A(PsiNorm)/(1+self.A(PsiNorm))))*(3*self.A(PsiNorm)/PsiNorm+2))
    #
    #def nu_1st_derivative(self,PsiNorm):
    #    return (self.nu(PsiNorm)/(1+self.A(PsiNorm)))*(3*self.A(PsiNorm)/PsiNorm + 2)
    #
    #def PSD_Eddington(self,r,s):
    #    EpsNorm = self.relativeEnergy(r,s)/self.energyNorm
    #    if EpsNorm < 0:
    #        return 0
    #    else:
    #        prefactor = 1./(np.sqrt(8)*self.energyNorm**(3./2.)*np.pi**2)
    #        integrand = lambda PsiNorm: self.nu_2nd_derivative(PsiNorm)/np.sqrt(np.abs(PsiNorm-EpsNorm))
    #        return prefactor*integrate.quad(integrand,0,EpsNorm)[0]
    
```

```python
rScale = 15*10**3 # Scale radius of 10 kpc
mwHaloDensity = 3.8 * 10 ** -2 # Solar masses per cubic parsec. This is the density that gives a mass of 10^12 for rScale = 10,000 kPc
delta = 200 # cutoff scale
criticalDensity = 1.26 * 10 ** -7 # Solar masses per cubic parsec

# Create an instance of the nfwHalo class
mwHalo = halo(rScale,mwHaloDensity,delta,criticalDensity)
print("This halo has a mass of " + "{:e}".format(mwHalo.haloMass) + " solar masses, and a concentration of " + str(mwHalo.c))
```

```python
nRadii = 200
nSpeeds = 200

maxRadius = mwHalo.rDelta
testradii = np.linspace(.001,maxRadius,100)
ms = np.zeros(100)
for k in range(100):
    ms[k] = mwHalo.maxSpeed(testradii[k])
maxSpeed = np.max(ms)

radii = np.linspace(0,maxRadius,nRadii) # in parsecs
speeds = np.linspace(0,maxSpeed,nSpeeds) # in km/s
```

```python
pdfEddingtonImage = np.zeros((nRadii,nSpeeds))
pdfMBImage = np.zeros((nRadii,nSpeeds))
relativeEnergies = np.zeros((nRadii,nSpeeds))

for i in range(nRadii):
    clear_output(wait=True)
    print(i/nRadii)
    for j in range(nSpeeds):
        pdfEddingtonImage[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_Eddington(radii[i],speeds[j])
        pdfMBImage[i,j] = (4*np.pi*radii[i]*speeds[j])**2*mwHalo.PSD_MB(radii[i],speeds[j])
        relativeEnergies[i,j] = mwHalo.relativeEnergy(radii[i],speeds[j])
        
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,9))
axis1 = ax1.imshow(np.transpose(pdfEddingtonImage),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
ax1.set_xticks(np.arange(0,nRadii,step_r))
ax1.set_xticklabels(np.round(radii[::step_r]))
ax1.set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
ax1.set_yticks(np.arange(0,nSpeeds,step_s))
ax1.set_yticklabels(np.round(speeds[::step_s]))
ax1.set_ylabel('Speed (km/s)')
ax1.set_title('Eddington NFW Phase Space')
fig.colorbar(axis1,ax=ax1)

ax1.contour(np.transpose(relativeEnergies),levels=[0],linewidths=3)

axis3 = ax2.imshow(np.transpose(pdfMBImage),origin='lower',aspect='auto',cmap='RdBu_r')

num_x_labels = 6
step_r = int(nRadii/(num_x_labels-1))
ax2.set_xticks(np.arange(0,nRadii,step_r))
ax2.set_xticklabels(np.round(radii[::step_r]))
ax2.set_xlabel('Radius (pc)')

num_y_labels = 7
step_s = int(nSpeeds/(num_y_labels-1))
ax2.set_yticks(np.arange(0,nSpeeds,step_s))
ax2.set_yticklabels(np.round(speeds[::step_s]))
ax2.set_title('Maxwell-Boltzmann NFW Phase Space')
ax2.set_ylabel('Speed (km/s)')
fig.colorbar(axis3,ax=ax2)

ax2.contour(np.transpose(relativeEnergies),levels=[0],linewidths=3)
```

```python
# Slicing at particular radii
fig, axs = plt.subplots(2,4,figsize=(36,18))

num_x_labels = 7
step_s = int(nSpeeds/(num_x_labels-1))

rIndices = .25*np.array((12,40,64,88,112,136,160,184))
print(rIndices)

for i in range(0,4):
    for j in range(0,2):
        k = 4*j+i
        axs[j,i].plot(pdfMBImage[int(rIndices[k]),:],label='Maxwell')
        axs[j,i].plot(pdfEddingtonImage[int(rIndices[k]),:],label='Eddington')
        axs[j,i].set_xticks(np.arange(0,nSpeeds,step_s))
        axs[j,i].set_xticklabels(np.round(speeds[::step_s]))
        axs[j,i].set_xlabel('Speed (km/s)')
        axs[j,i].set_title('Speed distribution at r = ' + str(np.round(radii[int(rIndices[k])])) + ' pc')
        axs[j,i].legend()
```

```python
nRadii = 20
maxRadius = mwHalo.rDelta

radii = np.linspace(0,maxRadius,nRadii) # in parsecs

massDensity_Eddington = np.zeros(nRadii)
massDensity = np.zeros(nRadii)

for i in range(nRadii):
    clear_output(wait=True)
    print(i/nRadii)
    massDensity_Eddington[i] = mwHalo.haloMass*integrate.quad(lambda x: (4*np.pi*x**2)*mwHalo.PSD_Eddington(radii[i],x),0,mwHalo.maxSpeed(radii[i]))[0]
    massDensity[i] = mwHalo.mass_density(radii[i])
```

```python
plt.figure()
plt.plot(radii,np.log10(massDensity_Eddington),label='Mass Density from Eddington')
plt.plot(radii,np.log10(massDensity),label='Mass Density')
plt.xlabel('Radius [pc]')
plt.ylabel('log10(mass density) [Solar mass/pc^3]')
plt.legend()

plt.figure()
plt.plot(radii,np.log10(np.abs(massDensity_Eddington-massDensity)),linewidth=5)
plt.xlabel('Radius [pc]')
plt.ylabel('Eddington - Exact [Solar mass/pc^3]')

```
